```python
# LOAD THE DATASET
import pandas as pd
import numpy as np

df = pd.read_csv('Downloads/Sales_data.csv')
print(df)
```

```python
# REMOVE DUPLICATES
df.drop_duplicates(inplace=True)
print(df)
```

```python
# CHANGE DATE TYPE
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
print(df)
```

```python
# ADD THE PROFIT_MARGIN COLUMN
df['Profit_Margin'] = (df['Profit'] / df['Sales']) * 100
print(df)
```

```python
# TOTAL SALES
Total_sales = df['Sales'].sum()
print(Total_sales)
```

```python
# TOTAL PROFIT
Total_profit = df['Profit'].sum()
print(Total_profit)
```

```python
# AVERAGE PROFIT MARGIN
Average_profit_margin = df['Profit_Margin'].mean()
print(Average_profit_margin)
```

```python
# REGION PROFIT
Region_profit = df.groupby('Region')['Profit'].sum()
print(Region_profit)
```

```python
# REGION SALES
Region_sales = df.groupby('Region')['Sales'].sum()
print(Region_sales)
```

```python
# CATEGORY SALES
Category_sales = df.groupby('Category')['Sales'].sum()
print(Category_sales)
```

```python
# TOP 10 PRODUCTS
Top_products = df.groupby('Product')['Sales'].sum().sort_values(ascending=True).head(10)
print(Top_products)
```

```python
# EXPORT THE CLEAN DATA CSV FILE
df.to_csv("clean_sales_data.csv", index=False)
```
